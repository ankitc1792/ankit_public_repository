### Step-4 : Splitting Data into Testing and Training Sets

```python
df_train,df_test = train_test_split(carprices,train_size=0.7,random_state=0)
```

>>Rescaling Features : Using MinMax scaler method in order to scale features<br> Features that require scaling are 'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price', 'symboling'

```python
scalingVariables = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price', 'symboling']
```

```python
scaler =MinMaxScaler()
```

```python
df_train[scalingVariables] = scaler.fit_transform(df_train[scalingVariables])
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (96, 60))
sns.heatmap(carprices.corr(),annot=True)
plt.show()
```

#### Dividing Data into X_train and y_train

```python
y_train = df_train.pop('price')
X_train = df_train
```

### Step 5 : Building the Model

#### Since we have a lot of variables for training purpose, therefore we using Automated Approach for Model Building<br>
>> Using Recursive Feature Elimination<br>Initially Taking 50 columns for RFE

```python
len(X_train.columns)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)

rfe = RFE(lm, 50)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

#### Building Model using statsmodel, for detailed statistics

```python
# Creating X_train_rfe for the usage with statsmodel
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
print(lm.summary())
```

> This is a multicollinearity Issue

> Columns below have very p values aranging from 0.066 to 0.618 <br>
[BrandName_audi, drivewheel_rwd, carheight, const, cylindernumber_five, BrandName_chevrolet, drivewheel_fwd, highwaympg,
horsepower, BrandName_isuzu, fuelsystem_spfi, BrandName_alfa-romero, stroke, cylindernumber_twelve, fuelsystem_mfi, 
cylindernumber_four, citympg]


> Let's also check VIF scores for these columns

```python
X_train_new = X_train_rfe.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### VIF scores does not look good at all in the first glance<br>
Taking 5% as baseline. The common columns with high VIF and high p values are<br>
These are the columns with High P value and High VIF therefore straight away remove these columns<br>
`['citympg', 'cylindernumber_four', 'cylindernumber_twelve','stroke','fuelsystem_spfi','horsepower','highwaympg','drivewheel_fwd'
 , 'cylindernumber_five', 'carheight', 'drivewheel_rwd']` <br>
 
 These are the columns with High p value and Low VIF therefore a further analysis is required<br>
 `['fuelsystem_mfi','BrandName_alfa-romero','fuelsystem_spfi','BrandName_isuzu', 'BrandName_chevrolet','BrandName_audi']`
 
 These are the columns with low p value and High VIF therefore a further analysis is required<br>
 `['enginetype_rotor', 'BrandName_porcshce', 'enginelocation_rear', 'cylindernumber_two', 'enginetype_ohcf', 'BrandName_peugeot', 'enginetype_l', 'BrandName_porsche','BrandName_subaru' ,'fuelsystem_idi' and so on ]`

Removing columns with High VIF and High p value

```python
col_to_drop = ['citympg', 'cylindernumber_four', 'cylindernumber_twelve','stroke','fuelsystem_spfi','horsepower','highwaympg','drivewheel_fwd', 'cylindernumber_five', 'carheight', 'drivewheel_rwd']
X_train_rfe2 = X_train_rfe.drop(col_to_drop,axis =1)
```

```python
lm = sm.OLS(y_train,X_train_rfe2).fit()
print(lm.summary())
```

>Earlier R-squared was 0.976 and Adj. R-squared was 0.965 which has now droped to <br>
R-squared = 0.968 and Adj. R-squared= 0.958<br> after droping many columns parameters haven't dropped much

> Also checking the VIF scores again

```python
X_train_new = X_train_rfe2.drop(['const'], axis=1)
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

>columns like ['compressionratio','fueltype_gas','fuelsystem_idi'] have high p value and High VIF so removing these columns

```python
col_to_drop = ['compressionratio','fueltype_gas','fuelsystem_idi']
X_train_rfe3 = X_train_rfe2.drop(col_to_drop,axis =1)
```

```python
lm = sm.OLS(y_train,X_train_rfe3).fit()
print(lm.summary())
```

```python
X_train_new = X_train_rfe3.drop(['const'], axis=1)
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Now is the Situation when we have either High VIF and Low Probability Or<br>
Low VIF and High Probability.
The approach that we will follow is by first removing columns with high p value--  This may also result in decreasing VIF values

```python
col_to_drop =['BrandName_alfa-romero','BrandName_toyota', 'BrandName_chevrolet', 'BrandName_isuzu','fuelsystem_mfi','BrandName_audi','BrandName_saab']
X_train_rfe4 = X_train_rfe3.drop(col_to_drop,axis =1)
```

```python
lm = sm.OLS(y_train,X_train_rfe4).fit()
print(lm.summary())
```

```python
X_train_new = X_train_rfe4.drop(['const'], axis=1)
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

> Now we have columns with controlled p value but VIF is still pretty bad, Removing columns one by one.<br>
>> After constantly removing columns one by one and checking out VIF, p-value, $R^{2}$, Adj  $R^{2}$ <br> We have finially stablized all the values

```python
list1 = ['enginetype_ohcf','BrandName_subaru','enginetype_l','BrandName_porsche','BrandName_porcshce','enginetype_rotor',
        'carlength','curbweight','carwidth','wheelbase','BrandName_peugeot','fuelsystem_2bbl','carbody_sedan','carbody_wagon',
        'boreratio','carbody_hatchback']
X_train_rfe5 = X_train_rfe4.drop(list1,axis =1)
```

```python
lm = sm.OLS(y_train,X_train_rfe5).fit()
print(lm.summary())
```

```python

X_train_new = X_train_rfe5.drop('const', axis=1)
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Step -6 : Residual Analysis of Training Data

> Residual Analysis of Training Data

>Checking if the Errors are normally distributed 

```python
y_train_price = lm.predict(X_train_rfe5)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

### Step - 7 : Making Prediciton with Final Model

```python
#Transforming the variables
df_test[scalingVariables] = scaler.transform(df_test[scalingVariables])
```

```python
df_test.describe()
```

>Dividing into X_test and y_test

```python
y_test = df_test.pop('price')
X_test = df_test
```

```python
X_test_new = sm.add_constant(X_test)
```

```python
X_test_new = X_test_new[X_train_rfe5.columns]
```

```python
# Making Prediction
y_pred = lm.predict(X_test_new)
```

### Step 8 : Evaluating Model

Plotting a graph between Actual and Predicted values

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)      
```
