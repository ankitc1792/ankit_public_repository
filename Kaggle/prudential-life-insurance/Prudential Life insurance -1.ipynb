```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('train.csv')
```

```python
df.info()
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
missing_value_df = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percent_missing})
missing_value_df.sort_values('percent_missing', inplace=True)
missing_value_df
```

```python
columns_to_remove  = list(missing_value_df[missing_value_df.percent_missing >10].column_name)
```

```python
df.drop(columns_to_remove,axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df_submission = pd.read_csv('test.csv')
```

```python
df_submission.drop(columns_to_remove,axis=1,inplace=True)
```

```python
df_submission[df_submission['Employment_Info_1'].isnull()]
```

```python
le = LabelEncoder()
df['Product_Info_2'] = le.fit_transform(df['Product_Info_2'])
```

```python
df_submission['Product_Info_2'] = le.transform(df_submission['Product_Info_2'])
```

```python
df.head()
```

```python
## submission na values
df_submit_add_later = df_submission[df_submission['Employment_Info_1'].isnull()]
df_submission.dropna(inplace=True)
```

```python
train_y = df['Response']
train_x = df.drop(columns=['Response'],axis=1)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
```

```python
model.fit(train_x,train_y)
```

```python
predict_train = model.predict(train_x)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_train = accuracy_score(train_y,predict_train)
print('\naccuracy_score on train dataset : ', accuracy_train)
```

```python
predict_submit = model.predict(df_submission)
```

```python
subdf =  pd.DataFrame(index=df_submission['Id'],data=predict_submit,columns=['Response'])
```

```python
toadd = pd.DataFrame(index=df_submit_add_later['Id'],data=[1,1,1],columns=['Response'])
```
